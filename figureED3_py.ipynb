```python
# fig S4C appearance times
matplotlib.rcParams['axes.linewidth'] = 0.2 

appearance_times = pd.read_csv('source_data/included/appearance_times_all.txt',sep='\t',index_col=0)

fig, ax = plt.subplots(figsize=(4,1.5))
i=0
colors=['k','#00AC00']
labels=['GFP-','GFP+']
for t in ['pmdtK_GFP-_20230310','pmdtK_GFP+_20230310']:
    data = appearance_times.loc[t,'appearance']
    count, bins_count = np.histogram(data, bins=100)
    pdf = count / sum(count)
    cdf = np.cumsum(pdf)
    ax.plot(bins_count[1:], cdf, color=colors[i],linewidth=3,label=labels[i])
    ax.set_ylim(0,1)
    ax.set_yticks([])
    ax.set_xlim([200,2200])
    i += 1
ax.tick_params(axis='both', which='major', labelsize=16)
ax.set_xlabel('Appearance Time, Minutes',size=16)  
matplotlib.rcParams['axes.linewidth'] = 1
plt.savefig('figures/figED3c_mdtK_sort_lag.pdf')
```
