```r
seurat <- readRDS('source_data/from_GEO/withCFT_ds_seurat_PCA_UMAP_clusters_v4.rds') # download from GEO 
seurat_archive <- seurat
```

```r
seurat@active.ident <- seurat$seurat_clusters
options(repr.plot.width=6, repr.plot.height=6)
my_cols = brewer.pal(10,"Spectral")
n_cells_per = 2000
alpha = 1
pt.size = 0.2
xmin = min(seurat$umap@cell.embeddings[,'UMAP_1'])-1
xmax = max(seurat$umap@cell.embeddings[,'UMAP_1'])+1
ymin = min(seurat$umap@cell.embeddings[,'UMAP_2'])-1
ymax = max(seurat$umap@cell.embeddings[,'UMAP_2'])+1

ids = levels(seurat$seurat_clusters)
print(ids)

sub_seurat=subset(seurat,idents=ids)
sampled.cells <- c()
set.seed(111)
for(id in levels(sub_seurat$seurat_clusters)){
    sampled.cells <- c(sampled.cells,sample(x = names(subset(sub_seurat,idents=id)$cell.type), size = n_cells_per, replace = F))
}

DimPlot(subset(sub_seurat,cells=sampled.cells),pt.size=pt.size,shuffle=TRUE) +
    xlim(xmin,xmax) + ylim(ymin,ymax) +
    scale_color_manual(name='Cell Clusters',labels = c("Early Stationary", "Stationary","Persister","Early Lag","Late Lag","Early Exponential","Late Exponential","Exponential CFT"), values = alpha(c(my_cols[c(1,2,4,7,8,9,10)],'black'),alpha))+ xlab('UMAP1') + ylab('UMAP2') +
    theme(text=element_text(family= "Arial", size=16),axis.text=element_text(size=16),legend.text=element_text(size=16),legend.position=c(0.7,0.85),legend.title = element_text(face = "bold"))
ggsave('figures/figED6g_withCFT_UMAP.pdf',width = 6, height = 6)
```

```r
seurat <- seurat_archive
seurat@active.ident <- seurat$cell.type

ids = c('CFT073_1hr_20231016')
print(ids)

levels(seurat$seurat_clusters) = c(levels(seurat$seurat_clusters),'10')
seurat@active.ident <- seurat$cell.type
sub_seurat=subset(seurat,idents=ids)
seurat$seurat_clusters[!names(seurat$seurat_clusters) %in% names(sub_seurat$seurat_clusters)] = '10'

seurat@active.ident <- seurat$cell.type
options(repr.plot.width=6, repr.plot.height=6)
my_cols = brewer.pal(10,"Spectral")
n_cells_per = 500
alpha = 1
pt.size = 1
xmin = min(seurat$umap@cell.embeddings[,'UMAP_1'])-1
xmax = max(seurat$umap@cell.embeddings[,'UMAP_1'])+1
ymin = min(seurat$umap@cell.embeddings[,'UMAP_2'])-1
ymax = max(seurat$umap@cell.embeddings[,'UMAP_2'])+1
sub_seurat$cell.type <- droplevels(sub_seurat$cell.type)
sampled.cells <- c()
set.seed(111)
for(id in levels(sub_seurat$cell.type)){
    sampled.cells <- c(sampled.cells,sample(x = names(subset(sub_seurat,idents=id)$cell.type), size = n_cells_per, replace = F))
}
seurat@active.ident <- seurat$seurat_clusters
sampled.cells <- c(sampled.cells,sample(x = names(subset(seurat,idents='10')$cell.type),size=10000,replace=FALSE))
alpha = c(0.1,1,1,1,1)
DimPlot(subset(seurat,cells=sampled.cells),pt.size=c(rep(2,n_cells_per),rep(1,10000)),order=rev(c('3','4','5','6','2','7'))) +
    xlim(xmin,xmax) + ylim(ymin,ymax) +
    scale_color_manual(labels = c('',"Late Lag","Early Exponential",'Late Exponential','Persister'),values = alpha(c('gray',my_cols[c(8,9,10,4)]),alpha)) + xlab('UMAP1') + ylab('UMAP2') +
    theme(text=element_text(family= "Arial", size=16),legend.position='none',legend.title = element_text(face = "bold")) +
    annotate("text",x=4, y=8, label='UPEC\n1 hour\n500 cells', size=16 / .pt, family="Arial")
ggsave('figures/figED6h_1hr.pdf',width = 6, height = 6)
```

```r
seurat <- seurat_archive
seurat@active.ident <- seurat$cell.type

ids = c('t4_1hr_WT_20210722','WT_1hr_20220331','WT_42min_20211217','d1_1hr_1_20220807','WT_1hr_20220527','tet2_20211014','wt_1hr_20231016')
wt_list = c()
for (ident in ids){
  #print(ident)
  prop = prop.table(as.integer(table(subset(seurat,idents=ident)$seurat_clusters)))[3]
  #print(prop)
  wt_list = c(wt_list,prop)
}

ids = c('CFT073_1hr_20231016') 
#print(ids)
CFT_list = c()
for (ident in ids){
  #print(ident)
  prop = prop.table(as.integer(table(subset(seurat,idents=ident)$seurat_clusters)))[3]
  #print(prop)
  CFT_list = c(CFT_list,prop)
}

ids = c('CFT073_amp_20231016')
#print(ids)
CFT_amp_list = c()
for (ident in ids){
  #print(ident)
  prop = prop.table(as.integer(table(subset(seurat,idents=ident)$seurat_clusters)))[3]
  #print(prop)
  CFT_amp_list = c(CFT_amp_list,prop)
}

ids = c('metG1_20200926','metG5_20201018_1','metG_52min_20211217')
metG_list = c()
for (ident in ids){
  prop = prop.table(as.integer(table(subset(seurat,idents=ident)$seurat_clusters)))[3]
  metG_list = c(metG_list,prop)
}

ids = c('tet1_20211014','tet5_20211014','tet1hr_20230919')
tet_list = c()
for (ident in ids){
  prop = prop.table(as.integer(table(subset(seurat,idents=ident)$seurat_clusters)))[3]
  tet_list = c(tet_list,prop)
}


fraction = c(wt_list,CFT_list,CFT_amp_list,metG_list,tet_list)
l = c("WT","CFT","CFT amp","metG*","tet")
type = rep(l, c(length(wt_list),length(CFT_list),length(CFT_amp_list),length(metG_list),length(tet_list)))
DATASET <- data.frame(fraction,type,stringsAsFactors = TRUE)

options(repr.plot.width=5, repr.plot.height=6)
set.seed(1)
DATASET$type <- factor(DATASET$type,levels=l)
my_comparisons <- list( l[1:2],l[c(1,3)] )

ggplot(DATASET, aes(type, 100*fraction)) +
    geom_boxplot(colour = my_cols[4],lwd=2,fatten=1) +
    geom_jitter(colour = 'black',size=2,width=0.1,height=0) + theme_classic()  + scale_y_log10() +
    theme(
        axis.title.x = element_blank(),
        text=element_text(family= "Arial", size=16),
        axis.text.y = element_text(family= "Arial", size=16),
        axis.text.x = element_text(family= "Arial", size=16),
        axis.title.y = element_text(family= "Arial", size=16),
        legend.position='none') + ylab('Cells in Persister Cluster (%)')

ggsave('figures/figED6j_boxplot.pdf',width = 4, height = 4)
```
